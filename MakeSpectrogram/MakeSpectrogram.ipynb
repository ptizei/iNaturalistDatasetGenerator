```python
from scipy.io import wavfile
import matplotlib.pyplot as plt
import numpy as np
```

```python
"10406.wav"
```

```python
# Read the wav file (mono)

samplingFrequency, signalData = wavfile.read("PardalRaven.wav")

# Plot the signal read from wav file
plt.figure(figsize=(10,20))
plt.subplot(211)


plt.title('Spectrogram of wav files with bird calls')


plt.specgram(signalData[:,1],Fs=samplingFrequency)

plt.xlabel('Time')

plt.ylabel('Frequency - PardalRaven')


plt.subplot(212) 
samplingFrequency, signalData = wavfile.read("Raven.wav")

plt.specgram(signalData[:],Fs=samplingFrequency)

plt.xlabel('Time')

plt.ylabel('Frequency - Raven')

```

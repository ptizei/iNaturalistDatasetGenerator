```python
from scipy.io import wavfile
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path
import pandas as pd
import seaborn as sns
import matplotlib
```

```python
matplotlib.rcParams['figure.figsize'] = [2.24, 2.24] # for square canvas
matplotlib.rcParams['figure.subplot.left'] = 0
matplotlib.rcParams['figure.subplot.bottom'] = 0
matplotlib.rcParams['figure.subplot.right'] = 1
matplotlib.rcParams['figure.subplot.top'] = 1

class Call:
    def __init__(self, sample_freq, signal_data):
        self.sample_freq = sample_freq
        self.signal_data = signal_data
        self.duration = len(signal_data)/sample_freq
    def plot_spectrogram(self, path, file, index):
        file_to_create = path / (file.stem+'-'+(str(index)+'.png'))
        #plt.figure(figsize=(2.24, 2.24))
        fig = plt.figure()
        ax=fig.add_subplot(1,1,1)
        plt.specgram(data_to_plot[:],Fs=cur_sample_freq)
        plt.axis('off')
        extent = ax.get_window_extent().transformed(fig.dpi_scale_trans.inverted())
        plt.savefig(file_to_create, bbox_inches=extent)
        plt.close(fig)

        

pathToData = Path('../Data/')  #change back to to Data to run the fullset      
window_dur = 10 #seconds
slide_step = 3  #seconds

if pathToData.exists:
    for item in pathToData.iterdir():
        if item.is_dir():
            pathToSubDir = pathToData / item.name
            for file in pathToSubDir.iterdir():
                if file.suffix == ".wav":
                    pathToFile = pathToSubDir / file.name
                    try:
                        cur_sample_freq, cur_signal_data = wavfile.read(pathToFile)
                    except:
                        pass
                    try:
                        if cur_signal_data.shape[1] > 1:
                            cur_signal_data = cur_signal_data[:,0] 
                    except:
                        pass
                    cur_call = Call(cur_sample_freq, cur_signal_data)
                    num_points_in_window = window_dur * cur_call.sample_freq
                    start_point = 0
                    end_point = start_point + num_points_in_window
                    i = 0
                    while end_point < len(cur_call.signal_data):
                        data_to_plot = cur_call.signal_data[start_point:end_point]
                        cur_call.plot_spectrogram(pathToSubDir, file, i)
                        start_point += (slide_step * cur_call.sample_freq)
                        end_point += (slide_step * cur_call.sample_freq)
                        print(start_point, end_point)
                        i += 1
                    print(i, 'files from', file.name,'in', pathToSubDir)
```

```python
from scipy.io import wavfile
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path
import pandas as pd
import seaborn as sns
```

```python
class Call:
    def __init__(self, sample_freq, signal_data):
        self.sample_freq = sample_freq
        self.signal_data = signal_data
        self.duration = len(signal_data)/sample_freq

pathToData = Path('../TestData/')        
        
if pathToData.exists:
    for item in pathToData.iterdir():
        if item.is_dir():
            pathToSubDir = pathToData / item.name
            for file in pathToSubDir.iterdir():
                if file.suffix == ".wav":
                    pathToFile = pathToSubDir / file.name
                    cur_sample_freq, cur_signal_data = wavfile.read(pathToFile)
                    cur_call = Call(cur_sample_freq, cur_signal_data)
                    
```

```python
cur_call.sample_freq
```

```python
window_dur = 10
slide_step = 3
num_points_in_window = window_dur * cur_call.sample_freq
start_point = 0
end_point = start_point + num_points_in_window

data_to_plot = cur_call.signal_data[start_point:end_point]
len(data_to_plot)
start_point += (slide_step * window_dur)
end_point += (slide_step * window_dur)

```

```python
# Read the wav file (mono)

samplingFrequency, signalData = wavfile.read("Raven.wav")

# Plot the signal read from wav file
plt.figure(figsize=(10,10))

fig = plt.figure()
ax=fig.add_subplot(1,1,1)
plt.specgram(signalData[:],Fs=samplingFrequency)
plt.axis('off')
extent = ax.get_window_extent().transformed(fig.dpi_scale_trans.inverted())
plt.savefig('raventest.png', bbox_inches=extent)
#plt.xlabel('Time')

#plt.ylabel('Frequency - Raven')

```
